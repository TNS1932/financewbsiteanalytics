```python
async function loadData() {
  const symbol = document.getElementById("symbol").value;

  const market = await fetch(`http://localhost:8000/market/${symbol}`);
  const marketData = await market.json();

  const personal = await fetch(`http://localhost:8000/portfolio/${symbol}`);
  const personalData = await personal.json();

  renderMarket(marketData);
  renderPersonal(marketData, personalData);
}

function renderMarket(data) {
  const labels = data.map(d => d.Date.split("T")[0]);
  const prices = data.map(d => d.Close);

  new Chart(document.getElementById("marketChart"), {
    type: "line",
    data: {
      labels,
      datasets: [{
        label: "Market Price",
        data: prices,
        borderWidth: 2
      }]
    }
  });
}

function renderPersonal(data, personal) {
  const labels = data.map(d => d.Date.split("T")[0]);
  const prices = data.map(d => d.Close);
  const avgLine = labels.map(_ => personal.avg_cost);

  document.getElementById("stats").innerHTML = `
    Shares: ${personal.total_shares}<br>
    Avg Cost: $${personal.avg_cost}<br>
    Price: $${personal.current_price}<br>
    Equity: $${personal.equity}<br>
    PnL: $${personal.pnl}<br>
    ROI: ${personal.roi_percent}%
  `;

  new Chart(document.getElementById("personalChart"), {
    type: "line",
    data: {
      labels,
      datasets: [
        {
          label: "Market Price",
          data: prices,
          borderDash: [5,5]
        },
        {
          label: "Your Avg Cost",
          data: avgLine,
          borderWidth: 2
        }
      ]
    }
  });
}
```
